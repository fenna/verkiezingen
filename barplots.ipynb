```python
import pandas as pd
data = pd.read_csv('uitslag.csv', sep =';')
data['percentage 2017'] = data['percentage 2017'].str.replace(',', '.').astype(float)
data['percentage 2021'] = data['percentage 2021'].str.replace(',', '.').astype(float)
df = data[['gemeente', 'percentage 2017', 'percentage 2021']]
df = df.set_index('gemeente')
df.head()
```

```python
drenthe = df[0:12]
flevoland = df[12:18]
fryslan = df[18:36]
gelderland = df[36:87]
groningen = df[87:97]
limburg = df[97:128]
brabant =  df[128:189]
noordholland = df[189:236]
overijssel = df[236:263]
utrecht= df[263:287]
zeeland = df[287:300]
zuidholland = df[300:353]
```

```python
def ax_plot(ax, title):   
    ax.plot.bar(rot=90, figsize=(15,3), ylim=(0,30), xlabel="", title=title).get_figure().savefig('images/'+title+'.jpg', bbox_inches='tight')
```

```python

ax_plot(groningen, 'Groningen')
ax_plot(drenthe, 'Drenthe')
ax_plot(fryslan, 'Fryslan')
ax_plot(gelderland, 'Gelderland')
ax_plot(overijssel, 'Overijssel')
ax_plot(flevoland, 'Flevoland')
ax_plot(zeeland, 'Zeeland')
ax_plot(noordholland, 'Noord-Holland')
ax_plot(zuidholland, 'Zuid-Holland')
ax_plot(brabant, 'Brabant')
ax_plot(limburg, 'Limburg')
ax_plot(utrecht, 'Utrecht')
```
